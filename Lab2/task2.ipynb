```python
#Working with dictionaries
#create dictionary 

student = {"name": "Maheshwar", "age": 21, "grade": "A"}
print(student)
```

```python
#Access value
student["name"]

```

```python

#use get()
student.get("age")
```

```python
#add item
student["city"] = "Mnr"
print(student)
```

```python
#update value
student["grade"] = "A+"
student["grade"]
```

```python
#remove item
del student["age"]
print(student)
```

```python
#remove item using pop()
student.pop("city") 
print(student)
```

```python
#remove last item
student.popitem()
print(student)
```

```python
#clear dictionary
student.clear()
print(student)
```

```python
#copy dictionary
dict = {"name": "Manu", "age": 21, "grade": "A+"}
new_student = dict.copy()
print(new_student)
```

```python
#get all keys
dict.keys()
```

```python
#get all values
dict.values()
```

```python
#get all items
dict.items()
```

```python
#check if key exists
"city" in dict
```

```python
#get length
len(dict)
```

```python
#loop through keys
for key in dict:
    print(key)  
```

```python
#loop through values
for value in dict.values():
    print(value)
```

```python
#loop through items
for key, value in dict.items():
    print(key, value)
```

```python
##Part 2:  Working with DataFrame
#Create a DataFrame containing student details such as Name, Age, and Marks.

import pandas as pd
df = pd.DataFrame({
    "Name": ["rohit", "mahesh", "Manisha","jayanti","dev","kalpesh"],
    "Age": [25, 21, 21,28,58,22],
    "Marks": [85, 90, 95,70,67,79]
})
print(df)
    


```

```python
#Display the first 5 rows of the DataFrame.
df.head()
```

```python
#display last 5 rows of the DataFrame
df.tail()
```

```python
#find total number of rows and columns in the DataFrame
df.shape
```

```python
#display all column names in the DataFrame
df.columns
```

```python
#display complete information about the DataFrame including data types and memory usage
df.info()
```

```python
#Generate a statistical summary of numerical columns in the DataFrame.
df.describe()
```

```python
#select and display the "Name" column from the DataFrame.
df["Name"]
```

```python
#select and display multiple columns such as Name and Marks
df[["Name", "Marks"]]
```

```python
#Display the first row of the DataFrame using positional indexing.
df.iloc[0]
```

```python
#Display the first row of the DataFrame using label-based indexing.
df.loc[0]
```

```python
#Access a specific cell value such as Marks of the first student.
df.loc[0, "Marks"]
```

```python
#Filter and display students whose Marks are greater than 80.
df[df["Marks"] > 80]    
```

```python
#Filter and display students whose Age is greater than 20 and Marks are greater than 75
df[(df["Age"] > 20) & (df["Marks"] > 75)]
```

```python
#Add a new column called Grade to the DataFrame.
df['Grade'] = ['A', 'A+', 'A+', 'B', 'C', 'B+']
print(df)
```

```python
#Increase all students' Marks by 5.
df["Marks"] = df["Marks"] + 5
print(df)
```

```python
#Update the Marks of Ram to 95 using a condition.
df.loc[df["Name"] == "rohit", "Marks"] = 95
print(df)
```

```python
#remove age column from the DataFrame
df.drop("Age", axis=1, inplace=True)
print(df)
```

```python
#remove the first row from the DataFrame
df.drop(0, axis=0)
```

```python
#sort the dataframe by Marks in ascending order
df.sort_values("Marks")
```

```python
#sort the DataFrame by Marks in descending order
df.sort_values("Marks", ascending=False)
```

```python
#concatenate two DataFrames vertically
df1 = pd.DataFrame({
    "Name": ["Alice", "Bob"],
    "Age": [25, 30],
    "Marks": [85, 90]
})
df2 = pd.DataFrame({
    "Name": ["Charlie", "David"],
    "Age": [35, 40],
    "Marks": [75, 80]
})
df = pd.concat([df1, df2], ignore_index=True)
print(df)
```
